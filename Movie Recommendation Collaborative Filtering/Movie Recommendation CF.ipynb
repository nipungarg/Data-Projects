```python
#Importing libraries

import pandas as pd
```

```python
#Importing datasets

movie_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
```

```python
#Merging datasets

df = movie_df.merge(ratings_df)
df.head()
```

## Cleaning and Preprocessing of data

```python
#Removing unnecessary columns

df.drop(["genres", "timestamp"], axis=1, inplace=True)
df.head()
```

```python
#Making matrix with user ratings in rows and movie titles in columns

df_new = pd.pivot_table(index="userId", columns="title", values="rating", data=df)
df_new.head()
```

```python
#Keeping only those columns which have atleast 10 non-null values

df_new.dropna(thresh=10,axis=1,inplace=True)

#Replacing NaN with 0's
df_new.fillna(0, axis=1, inplace=True)
```

```python
df_new.head()
```

Finding similarities using Pearson Correlation

```python
df_movies = df_new.corr(method="pearson")
df_movies.head()
```

```python
#Creating a function block to find similar movies based on ratings given by a user

def similar_movies(movie_name, rating):
    similar_ratings = df_movies[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    
    return similar_ratings
```

```python
#User ratings to some movies provided, showing which genre she likes

romantic_lover = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3)
                  ,("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]


#Creating a new dataframe with similar movies
similar = pd.DataFrame()

for movie,rating in romantic_lover:
    similar = similar.append(similar_movies(movie, rating), ignore_index=True)
    
    
similar.head()
```
